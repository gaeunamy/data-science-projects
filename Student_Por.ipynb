```python
import pandas as pd

# 데이터셋 불러오기
df = pd.read_csv('/content/drive/MyDrive/DS Project 2024/week04/student-lpor.csv')

# 데이터의 기본 정보 확인
print("데이터셋의 기본 정보:")
print(df.info())

# 데이터의 처음 5개 행 확인
print("\n데이터의 처음 5개 행:")
print(df.head())

# 기술통계량 확인
print("\n데이터의 기술통계량:")
print(df.describe())

# 각 열의 결측치 개수 확인
print("\n각 열의 결측치 개수:")
print(df.isnull().sum())

# 각 열의 데이터 타입 확인
print("\n각 열의 데이터 타입:")
print(df.dtypes)

# 범주형 열의 고유값 확인
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column} 열의 고유값:")
    print(df[column].value_counts())

# 수치형 열의 분포 확인
numerical_columns = df.select_dtypes(exclude=['object']).columns
for column in numerical_columns:
    print(f"\n{column} 열의 분포:")
    print(df[column].describe())

# 시각화를 통한 데이터 탐색
import seaborn as sns
import matplotlib.pyplot as plt

# 수치형 열의 분포 시각화
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 3, i+1)
    sns.histplot(data=df, x=column, kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()

# 범주형 열의 분포 시각화
plt.figure(figsize=(12, 6))
for i, column in enumerate(categorical_columns):
    plt.subplot(2, 3, i+1)
    sns.countplot(data=df, x=column)
    plt.title(column)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 등급을 부여하는 함수
def grade(score):
    if score >= 15:
        return 'A'
    elif score >= 10:
        return 'B'
    elif score >= 5:
        return 'C'
    else:
        return 'D'

# 등급 부여
df['G1_grade'] = df['G1'].apply(grade)
df['G2_grade'] = df['G2'].apply(grade)

# G1 등급별 비율
G1_grade_counts = df['G1_grade'].value_counts()
G1_grade_labels = G1_grade_counts.index
G1_grade_sizes = G1_grade_counts.values

# G2 등급별 비율
G2_grade_counts = df['G2_grade'].value_counts()
G2_grade_labels = G2_grade_counts.index
G2_grade_sizes = G2_grade_counts.values

# pie chart 그리기
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.pie(G1_grade_sizes, labels=G1_grade_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of G1 Grades')

plt.subplot(1, 2, 2)
plt.pie(G2_grade_sizes, labels=G2_grade_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of G2 Grades')

plt.tight_layout()
plt.show()
```
